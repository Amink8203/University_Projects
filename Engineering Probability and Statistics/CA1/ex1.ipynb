```python
import numpy as np
import matplotlib.pyplot as plt
def comp_binomial(n,m, prob):
    arr = np.random.choice([0,1], size = (m * n) , p = [1 - prob, prob])
    two_d_arr = arr.reshape(m, n)
    row_sums = np.sum(two_d_arr, axis=1)
    return row_sums

p_range = np.arange(0.01,1.01,0.01)
var_theory = []
var_prac = []
mean_theory = []
mean_prac = []
n = 500
m = 5000
for p in p_range:
    q = 1 - p
    result = comp_binomial(n,m,p)
    mean_theory.append(n*p)
    mean_prac.append(np.mean(result))
    var_prac.append(np.var(result))
    var_theory.append(n*p*q)

plt.subplot(221)
plt.plot(p_range,mean_theory,color = "tab:blue", label = "mean_theoritical")
plt.plot(p_range,mean_prac,color = "tab:red", label = "mean_practical")
plt.legend()
plt.title("Mean_plot")
plt.subplot(222)
plt.plot(p_range,var_theory,color = "tab:blue", label = "var_theoritical")
plt.plot(p_range,var_prac,color = "tab:red", label = "var_practical")
plt.title("Var_plot")
plt.legend()
plt.show()
plt.subplot(223)
plt.plot(p_range,mean_theory,color = "tab:blue", label = "mean_theoritical")
plt.scatter(p_range,mean_prac,color = "tab:red", label = "mean_practical")
plt.title("Mean_scatter")
plt.legend()
plt.subplot(224)
plt.plot(p_range,var_theory,color = "tab:blue", label = "var_theoritical")
plt.scatter(p_range,var_prac,color = "tab:red", label = "var_practical")
plt.title("Var_scatter")
plt.legend()
plt.show()


```
