```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
from scipy.stats import norm

n = 250
p = 0.008
x_ax = np.arange(0,11,1)
bino = np.random.binomial(n,p,10000)
bino_probs = [0] * 11
for num in bino:
    if num >= 0 and num <= 10:
        bino_probs[num] += 1
    else:
        continue
sum = 0
for num in bino_probs:
    sum += num
for i in range(len(bino_probs)):
    bino_probs[i] = bino_probs[i] / sum

poi_landa = 2

poi_probs = []
for i in range(11):
    poi_probs.append(poisson.pmf(k = i, mu = poi_landa))


mean = 2
var = 1.984
s_deviation = math.sqrt(var)

norm_probs = []
norm_distro = norm(loc = mean,scale = s_deviation)
for i in range(11):
    norm_probs.append(norm_distro.pdf(x=i))

plt.plot(x_ax,bino_probs,color = "tab:red", label="binomial")
plt.plot(x_ax,poi_probs,color = "tab:blue", label="poisson")
plt.plot(x_ax,norm_probs,color = "tab:green", label="normal")
plt.legend()
plt.show()
```
