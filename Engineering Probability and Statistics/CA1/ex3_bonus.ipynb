```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

n = 200
p = 0.65
poi_landa = n * p
exponential_landa = 30
samples = 100000
uniform_samples = np.random.uniform(0, 100, samples)
poisson_samples = np.random.poisson(poi_landa,samples)
exponential_samples = np.random.exponential(exponential_landa,samples)
sum_of_scores = uniform_samples + poisson_samples + exponential_samples
#showing histogram of sum of scores
plt.hist(sum_of_scores, density=True, color="red", edgecolor="black", bins = 100, label= "sums histogram")
plt.legend()

normal_s_deviation = np.std(sum_of_scores)
normal_mean = np.mean(sum_of_scores)
normal_distro = np.random.normal(loc=normal_mean, scale=normal_s_deviation,size=samples)
#using limit for left and right side of normal graph that is approximately 0
left, right = plt.xlim()
x_ax = np.arange(left,right,0.1)
plt.plot(x_ax,norm.pdf(x_ax, loc = normal_mean, scale = normal_s_deviation),label="normal plot")
plt.legend()
plt.show()
# sum of scores of the three lessons is x_axis
```
