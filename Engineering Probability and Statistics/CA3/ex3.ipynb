```python
# Change pace and dribbling columns and update dataframe(Question1)
import pandas as pd
import math
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api
import scipy.stats as stats
import random
def set_seed(seed):
    np.random.seed(seed)
    random.seed(seed)
set_seed(810109203)
df = pd.read_csv('FIFA2020.csv', encoding="ISO-8859-1")
df['pace'] = pd.to_numeric(df['pace'], errors='coerce')
mean_pace = df['pace'].mean()
mean_pace = math.ceil(mean_pace)
df['pace'].fillna(mean_pace, inplace=True)

df['dribbling'] = pd.to_numeric(df['dribbling'], errors='coerce')
mean_dribbling = df['dribbling'].mean()
mean_dribbling = math.ceil(mean_dribbling)
df['dribbling'].fillna(mean_dribbling, inplace=True)
```

```python
# Calculating quartiles for age and drawing box plot(Question 2)
plt.boxplot(df['age'].dropna())
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.show()

Q1 = df['age'].quantile(0.25)
Q2 = df['age'].median()
Q3 = df['age'].quantile(0.75)
min_age = df['age'].min()
max_age = df['age'].max()

print(f"Minimum Age: {min_age}")
print(f"Q1 (First Quartile): {Q1}")
print(f"Q2 (Second Quartile or Median): {Q2}")
print(f"Q3 (Third Quartile): {Q3}")
print(f"Maximum Age: {max_age}")
```

```python
# Question 3 (part A and B)
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')

n = 100
samples = np.random.choice(df['weight'].dropna(), size=n, replace=False)

mean = np.mean(samples)
sd = np.std(samples)
variance = np.var(samples)

print(f"Mean of {n} Random Samples: {mean}")
print(f"Standard Deviation of {n} Random Samples: {sd}")
print(f"Variance of {n} Random Samples: {variance}")
```

```python
# Question 3 (part D)
# Compare normal distribution with sample distribution
normal_distribution = np.random.normal(loc=mean, scale=sd, size=n)
fig, ax = plt.subplots(figsize=(8, 8))
statsmodels.api.qqplot_2samples(samples, normal_distribution, line='45', ax=ax)
ax.set_title("Q-Q Plot for Two Distributions")
plt.show()
```

```python
# Shapiro-Wilk test
statistic, p_value = stats.shapiro(samples)
if p_value > 0.05:
    print(f"Since p-value({p_value}) is greater than {0.05} ,the sample distribution is normal.(Fail to reject the null hypothesis)")
else:
    print(f"Since p-value({p_value}) is smaller than {0.05} ,the sample distribution is not normal.(Reject the null hypothesis)")
```
